```python
#Vamos importar as bibliotecas

import pandas as pd
import seaborn as sns
```

```python
#Vamos analisar primeiro a base de varejo

vendas = pd.read_excel('varejo.xlsx')

vendas.head()
```

```python
#Vamos analisar mais informaes

vendas.info()
```

```python
#Repare que para preo e estado possuem dados nulos. Vamos tratar esses dados nulos posteriormente

#Vamos analisar os valores dessas colunas agora utilizando o unique para ver a quantidade de categorias de cada um

vendas.groupby('idcanalvenda').idcompra.nunique() #Quais os tipos diferentes de idcanalvendas com idcompra unico, isso vai dar, por exemplo, 139 idcompra dentro do da categoria app do idcanalvenda
```

```python
#Vamos agrupar APP com Aplicativo para ficarem na mesma categoria utilizando a funo replace

vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
```

```python
#Vamos verificar como ficou:

vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
#Vamos verificar agora as outras categorias:

#bandeira
vendas.groupby('bandeira').idcompra.nunique()
```

```python
#Nome_Departamento
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
#estado

vendas.groupby('estado').idcompra.nunique()
```

```python
#Percebemos que nas categorias de Nome_departamento temos o nome de algumas categorias sem o _, o que no  bom deixar. Ento vamos corrigir isso com o replace:

#Conseguimos substituir de uma vez s todos os espaos em branco pelo _ sem precisar ir de um por um
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ', '_')

#Vamos ver como ficou:
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
#Vamos tratar agora os dados nulos

#Vimos la em cima que preo e estado possuem menos dados que os demais, ou seja, temos dados nulos

#Vamos ver apenas os estados nulos:

vendas[vendas['estado'].isnull()]
```

```python
#Temos algumas formas de tratar esses dados nulos

#Uma delas  eliminando a linha inteira onde possui o dado nulo

vendas_sem_nulos = vendas.dropna()

vendas_sem_nulos.info()
```

```python
#Porm, na premissa de negcio pede pra colocar MS onde no tiver estado

#iremos utilizar a funo .fillna e o inplace

vendas['estado'].fillna('MS', inplace=True) #O inplace serve pra substituir na coluna oficial

#Vamos ver como ficou agora

vendas.info()
```

```python
#Vamos ver agora como faremos pra preo

#Uma das maneiras  substituir os valores pela mdia dos preos:

#Calculando a media
media_preco = vendas['Preo'].mean()

vendas['Preo'].fillna(media_preco, inplace = True)

#Vamos ver como ficou agora

vendas.info()
```

```python
#Outra forma de bustituir os nulos do preo poderia ser com a mediana

mediana_preco = vendas['Preo'].median()
mediana_preco
```

```python
#Ou por qualquer outro valor que eu queira

preco_desejado = 1000 #Exemplo
```

```python
#Vamos analisar agora os valores da base de dados

vendas[['Preo', 'Preo_com_frete']].describe()
```

```python
#Temos outra premissa de negcio de que o preo no pode ser maior que o preo com frete

#Eu posso fazer uma nova base filtrando os valores ou posso eliminar os valores na minha base existente

preco_errado = vendas.query('Preo > Preo_com_frete')

vendas_correto = vendas.query('Preo < Preo_com_frete')

vendas_correto.describe()
```

```python
#Vamos relembrar alguns filtros

vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
#Vamos ver agora os departamentos mais vendidos

vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
#Vamos ver a media de preos com frete por nome de departamento:

round(vendas_correto.groupby("Nome_Departamento")['Preo_com_frete'].agg('mean').sort_values(ascending = False).reset_index(),2) #Utilizando o round pra deixar duas casas decimais
```

```python
#Vendo agora a quantidade de vendas por ms

#Por data
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending = False)
```

```python
#Por ms

#Vamos criar dentro do prprio dataset uma coluna pra extrair o ms

vendas_correto['mes'] = vendas_correto['Data'].dt.month

vendas_correto
```

```python
#Agora fazendo por ms

vendas_correto.groupby('mes').idcompra.nunique().sort_values(ascending = False)
```

```python
#Vamos importar agora a base cliente_varejo

cliente = pd.read_excel('cliente_varejo.xlsx')

cliente.head()
```

```python
#Vamos analisar a base com o describe

cliente.describe()
```

```python
#Analisando com o info

cliente.info()
```

```python
#Vamos transformar a renda no float

cliente = cliente.astype({'renda':'float'})

cliente.info()
```

```python
#Vamos juntar as duas tabelas

vendas_cliente = vendas_correto.merge(cliente, how = 'left', on = ['cliente_Log'])

vendas_cliente.head()
```

```python
#Vamos fazer a mdia de renda para cada tipo de canal de venda

agg_idcanal_renda = round(vendas_cliente.groupby(by = ['idcanalvenda']).agg({'renda':'mean'}).sort_values(by = ['renda'], ascending = False).reset_index(),2)

agg_idcanal_renda
```

```python
#Vamos fazer agora a mdia de idade de clientes por bandeira

agg_idade_bandeira = round(vendas_cliente.groupby(by = ['bandeira']).agg({'idade':'mean'}).sort_values(by = ['idade'], ascending = False).reset_index(),2)

agg_idade_bandeira
```

```python
#vamos plotar este resultado pelo matplotlib

import matplotlib.pyplot as plt
```

```python
#Setando o tamanho do grfico

fig = plt.figure(figsize = (7, 4))
```

```python
#Plotando um grfico de barras

plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color = 'green')

plt.ylabel('Mdia de idade')

plt.title('Idade mdia por Bandeira', loc = 'center', fontsize = 18, fontweight = 'bold', color = 'black');
```

```python
#Fazendo para o idcanalvenda

plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color = 'green')

plt.ylabel('Mdia de renda')

plt.title('Renda pelo canal de venda', loc = 'center', fontsize = 18, fontweight = 'bold', color = 'black');
```

```python
#Vamos fazer uma representao por data agora:

venda_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()

venda_por_data
```

```python
#Fazendo um grfico de linha

fig = plt.figure(figsize = (20,12))

plt.xlabel('Data da venda')
plt.ylabel('Quantidade de vendas')

x = venda_por_data['Data']
y = venda_por_data['idcompra']

plt.plot(x, y, color = 'green', linewidth = '2')

plt.show()
```

```python
#Vamos fazer com a mdia de preos por nome  de departamento

agg_dept_preco = round(vendas_correto.groupby("Nome_Departamento")['Preo_com_frete'].agg('mean').sort_values(ascending = False).reset_index(),2)

fig = plt.figure(figsize = (20,12))

plt.xlabel('Departamento')
plt.ylabel('Mdia de preo com frete')
plt.title('Mdia de Preo por departamento', loc = 'center', fontsize = 18, fontweight = 'bold', color = 'black');

x = agg_dept_preco['Nome_Departamento']
y = agg_dept_preco['Preo_com_frete']

plt.plot(x, y, color = 'green', linewidth = '2')

plt.show()
```
